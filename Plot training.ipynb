```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("training_info.csv")
df_ = pd.read_csv("training_info_.csv")
```

```python
df["DQN-LunarLander"] = df_["DQN-LunarLander"]
df["Episodes"] = [i for i in range(df.shape[0])]
print(df.shape)
df.head(5)
```

```python
plt.figure(figsize=(15,5))
plt.title("LunarLander-v2 (Reward vs Episodes)")
plt.plot(df["DuelingDQN-LunarLander"], label = "Dueling DQN", color = "blue")
plt.plot(df["DQN-LunarLander"], label = "DQN", color = "red")
plt.ylabel("Reward/Score")
plt.xlabel("Episode")
plt.legend()
plt.show()
```

```python
scores_dueling_dqn = list(df["DuelingDQN-LunarLander"])
y_dueling_dqn = [np.array(scores_dueling_dqn[max(0,i-100):i+1]).mean() for i, x in enumerate(scores_dueling_dqn)]

scores_dqn = list(df["DQN-LunarLander"])
y_dqn = [np.array(scores_dqn[max(0,i-100):i+1]).mean() for i, x in enumerate(scores_dqn)]
```

```python
plt.figure(figsize=(15,5))
plt.title("LunarLander-v2 (Avg. Reward vs Episodes)")
plt.plot(y_dueling_dqn, label = "Dueling DQN", color = "blue")
plt.plot(y_dqn, label = "DQN", color = "red")
plt.ylabel("Avg. Reward/Score")
plt.xlabel("Episode")
plt.legend()
plt.show()
```
